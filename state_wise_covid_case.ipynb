```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup 
```

```python
url = "https://www.mohfw.gov.in"
```

```python
data = requests.get(url)
data
```

```python
data_text = data.text
data_text = data_text.replace("<!--<tbody>","")
data_text = data_text.replace("<tr></tr>","")
data_record = BeautifulSoup(data_text,"html5lib")
```

```python
data_record_table = data_record.find_all('table')
```

```python
data_record_thead = data_record.find_all('tbody')
data_record_row = data_record_thead[0].find_all('tr')
```

```python
covid_data1.empty
covid_data1 = pd.DataFrame(columns=["S_No","Active_Cases","Cured_Discharged_Migrated","Deaths","State "])
covid_data1
```

```python
for row in (data_record_row):
    col = row.find_all('td')
    if (col[0].text != ''):
        print(col)
        if (col!=[]):
            S_No = col[0].text
            Name_of_State = col[1].text
            Active_Cases = col[2].text
            Cured_Discharged_Migrated = col[3].text
            Deaths = col[4].text
            covid_data1 = covid_data1.append({"S_No":S_No,"Active_Cases":Active_Cases,"Cured_Discharged_Migrated":Cured_Discharged_Migrated,
                                              "Deaths":Deaths,"State":Name_of_State},ignore_index=True)
```
